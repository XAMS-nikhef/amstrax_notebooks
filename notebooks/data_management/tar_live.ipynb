# Mass tar the live data into `.tar`
Live data is not so nice for a file system as there are many many files.

In this notebook we tar live-data folders to minimize the number of files on the file system.

december 2021

```python
import amstrax
import os
```

```python
st = amstrax.contexts.xams_little()
```

```python
runs = st.select_runs()
```

```python
runs_done = list(runs['name'][runs['raw_records_v1730_available']])
runs_done
```

```python
live_dir = '/data/xenon/xamsl/live_data/'
from strax.utils import tqdm
import time
```

Below we loop over the runs and make the tar files

```python
for run in tqdm(runs_done):
    path = os.path.join(live_dir, run)
    assert os.path.exists(path)
    tar_path = f'{path}.tar.gz'
    if os.path.exists(tar_path):
        cmd = f'rm -r {path}'
        print(cmd)
    else:
        cmd = f'tar -zcf {tar_path} {path}'
        print(cmd, end='\r', flush=True)
    !$cmd
    time.sleep(1)
```
